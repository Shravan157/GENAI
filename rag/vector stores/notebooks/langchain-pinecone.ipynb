```python
import os 
from langchain_huggingface import HuggingFaceEndpointEmbeddings
from langchain_pinecone import PineconeVectorStore
from dotenv import load_dotenv
from langchain_core.documents import Document
```

```python
load_dotenv()

embedding_model = HuggingFaceEndpointEmbeddings(
    model = "sentence-transformers/all-MiniLM-L6-v2"
)
```

```python

doc1 = Document(
    page_content="Apple is a sweet and crunchy fruit. It is commonly red, green, or yellow in color and is often eaten fresh or used in juices, pies, and salads.",
    metadata={
        "fruit_name": "Apple",
        "color": "Red/Green/Yellow",
        "taste": "Sweet",
        "category": "pome fruit",
        "source": "fruits_doc_1",
    },
)

doc2 = Document(
    page_content="Mango is a tropical fruit known for its juicy texture and rich sweet flavor. It is often called the king of fruits and is widely used in shakes, desserts, and pickles.",
    metadata={
        "fruit_name": "Mango",
        "color": "Yellow/Orange",
        "taste": "Sweet",
        "category": "tropical fruit",
        "source": "fruits_doc_2",
    },
)

doc3 = Document(
    page_content="Banana is a soft and sweet fruit with a yellow peel. It is commonly eaten as a quick snack and is often added to smoothies, cereals, and desserts.",
    metadata={
        "fruit_name": "Banana",
        "color": "Yellow",
        "taste": "Sweet",
        "category": "tropical fruit",
        "source": "fruits_doc_3",
    },
)

doc4 = Document(
    page_content="Orange is a citrus fruit known for its juicy pulp and refreshing tangy taste. It is commonly consumed fresh or used to make orange juice.",
    metadata={
        "fruit_name": "Orange",
        "color": "Orange",
        "taste": "Tangy",
        "category": "citrus fruit",
        "source": "fruits_doc_4",
    },
)

doc5 = Document(
    page_content="Grapes are small round fruits that grow in bunches. They can be green, black, or purple and are used to make juice, raisins, and jams.",
    metadata={
        "fruit_name": "Grapes",
        "color": "Green/Black/Purple",
        "taste": "Sweet/Sour",
        "category": "berry fruit",
        "source": "fruits_doc_5",
    },
)
```

```python
docs = [doc1,doc2,doc3,doc4,doc5]
```

```python
## creating the vector store 

vector_store = PineconeVectorStore.from_documents(
    embedding=embedding_model,
    documents=docs,
    index_name = 'langchainvector'
)
```

```python
## fetching the data from the vector store 
index = vector_store.get_pinecone_index(index_name='langchainvector')
```

```python
result = index.fetch(
    ids=['14bdb37c-756f-4366-94f3-99ae33ebc63c','857ab144-73fe-441c-8891-f742492eaddb'],
    namespace=""
)

result
```

```python
result
```

#### fetching items from the documents

```python
for vector_id,vector_data in result.vectors.items():
    metadata = vector_data.metadata
    print(f'ID: {vector_id}')
    print(f'Text: {metadata['text']}')
    print(f'Metadata: {metadata}')
    print(f'Embeddings: {vector_data.values}')
    print(f'Embedding_dimensions: {len(vector_data.values)}')
    print('-'*30)
```

#### doing similarity search

```python
vector_store.similarity_search(
    query= 'which fruit is the king',
    k = 1
)
```

#### doing similarity search with score

```python
vector_store.similarity_search_with_score(
    query= 'which fruit is the king',
    k = 1
)
```

#### filtering the records

```python
vector_store.similarity_search_with_score(
    query= 'which fruit is the king',
    metadata= {'metadata.category' : 'tropical fruit'},
    k =2
)
```

#### updating document
#### case 1 : metadata change

```python
updated_index = index.update(
    id="857ab144-73fe-441c-8891-f742492eaddb",
    set_metadata={
        "taste" : "sometime tangy sometime sweet, rich with vitamin c"
    },
    namespace= ""
)
## this thing only updates the metadata, does not change the text and the embedding values
```

```python
result_updated = index.fetch(
    ids=['857ab144-73fe-441c-8891-f742492eaddb'],
    namespace= ""
)

for vector_id , vector_data in result_updated.vectors.items():
    metadata = vector_data.metadata
    print(f'ID: {vector_id}')
    print(f'updated metadata: {metadata['taste']}')
```

#### case 2 : document text change

```python
## we are trying to update the text and the metadata for grapes 

vector_id = "14bdb37c-756f-4366-94f3-99ae33ebc63c"

updated_text = "Grapes are small berry fruits known for their sweet and slightly tangy taste."

updated_metadata = {
    "category": "citrus fruit",
    "color": "Pink/Red/White",
    "fruit_name": "Grapefruit",
    "source": "fruits_doc_5",
    "taste": "Bitter/Tangy",
    "text": updated_text
}

update_embedding = embedding_model.embed_query(updated_text)

index.update(
    id=vector_id,
    values=update_embedding,
    namespace="",
    set_metadata=updated_metadata
)
```

```python

result_updated = index.fetch(
    ids=['14bdb37c-756f-4366-94f3-99ae33ebc63c'],
    namespace= ""
)

for vector_id , vector_data in result_updated.vectors.items():
    metadata = vector_data.metadata
    print(f'ID: {vector_id}')
    print(f'updated metadata: {metadata}')
    print(f'updated text: {metadata['text']}')
    print(f'updated embeddings: {vector_data.values[0:5]}') # fetching 5 embeddings
```

```python
## deleting items

index.delete(
    ids=['857ab144-73fe-441c-8891-f742492eaddb'],
    namespace=""
)
```
